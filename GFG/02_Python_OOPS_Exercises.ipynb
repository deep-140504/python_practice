#### 1. Python program to build flashcard using class in Python

```python
class FlashCard():
    def __init__(self, word, meaning):
        self.word = word
        self.meaning = meaning
    def __str__(self):
        # it is required to add self here as an argument, otherwise it wont work
        return self.word + ' ( ' + self.meaning + ' ) '
```

```python
flash = []
while(True):
    word = input("input the word you want a flashcard for: ")
    # meaning = input("input the meaning of ", word, ": ")
    # In Python, the input() function takes only one string argument,
    # and you cannot directly concatenate variables inside it. Instead, you should use string formatting.
    # meaning = input(f"Input the meaning of {word}: ") this can be done
    meaning = input("input the meaning of word: ")
    flash.append(FlashCard(word, meaning))

    exit = int(input("If you want to continue adding flashcards, enter 0: "))
    if(exit):
        break
for i in flash:
    print("< ", i)
```

#### 2. Shuffle a deck of card with OOPS in Python

```python
import random
suites = ("Hearts", "Diamonds", "Spades", "Clubs") # here the values are constant so prefer using tuples, instead of lists
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')

class Card:
    def __init__(self, rank, suite):
        self.rank = rank
        self.suite = suite
    def __str__(self):
        return (f"{self.rank} of {self.suite}")
class Deck:
    def __init__(self):
        self.all_cards = []
        for suite in suites:
            for rank in ranks:
                self.all_cards.append(Card(rank, suite))
    def display(self):
        for i in range(len(self.all_cards)):
            print(self.all_cards[i])

    def shuffle(self):
        random.shuffle(self.all_cards)

curr_deck = Deck()
curr_deck.display()
curr_deck.shuffle()
print()
curr_deck.display()
```

#### 4. Student management system in Python

```python
class Student:
    def __init__(self, name, rn, m1, m2):
        self.name = name
        self.rn = rn
        self.m1 = m1
        self.m2 = m2

    def accept(self, name, rn, m1, m2):
        ob = Student(name, rn, m1, m2)
        ls.append(ob)
        # the thing to take into consideration is that, the value is accepted through the accept function and not through the constructor
        # here, the same is done, as if the logic of appending the newly created object is combined inside of the class
        # so, let's say if the object being was appended outside of the class then it can be very clumsy and on top, it will cause repetition of the code

    # Display student information
    def display(self):
        print("Name: ", self.name)
        print("Roll No: ", self.rn)
        print("Mark1: ", self.m1)
        print("Mark2: ", self.m2)
        print()

    # Search for a student by roll number
    def search(self, rn):
        for i in range(len(ls)):
            if ls[i].rn == rn:
                return i
        return -1  # return -1 if student not found

    # Delete student by roll number
    def delete(self, rn):
        i = self.search(rn)
        if i != -1:  # Ensure the student exists
            del ls[i]

    # Update a student's roll number
    def update(self, rn, new_roll_no):
        i = self.search(rn)
        if i != -1:  # Ensure the student exists
            ls[i].rn = new_roll_no  # Update the roll number

ls = []

s1 = Student("Alice", 101, 85, 90)
s2 = Student("Bob", 102, 75, 80)

s1.accept("Charlie", 103, 95, 92)
s2.accept("David", 104, 88, 84)

for student in ls:
    student.display()

s1.update(101, 111)

print("Updated list of students:")
for student in ls:
    student.display()

s1.delete(104)

print("Final list after deletion:")
for student in ls:
    student.display()
```
