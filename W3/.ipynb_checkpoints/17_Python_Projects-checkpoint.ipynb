#### 4. Create a Python project that prints out every line of the song "99 bottles of beer on the wall.

```python
def fun():
    n = 99
    while True:
        s1 = f"{n} bottles of beer on the wall, {n} bottles of beer."
        s2 = f"Take one down and pass it around, {n - 1} bottles of beer on the wall."
        print(s1)
        print(s2)
        n = n - 1
        if(n == 0):
            a = "No more bottles of beer on the wall, no more bottles of beer."
            b = "Go to the store and buy some more, 99 bottles of beer on the wall."
            print(a)
            print(b)
            break

fun()
```

#### 1. Create a Python project to get the value of Pi to n number of decimal places.

```python
# calculation is carried out using Chudnovsky Algorithm
import math
from decimal import Decimal, getcontext

def calculate_pi(decimal_places):
    getcontext().prec = decimal_places + 2
    constant_c = 426880 * Decimal(math.sqrt(10005))
    numerator_term = 1
    linear_term = 13591409
    exponential_term = 1
    factorial_term_increment = 6
    summation = linear_term

    for iteration in range(1, decimal_places):
        numerator_term = numerator_term * (factorial_term_increment**3 - 16 * factorial_term_increment) // iteration**3
        linear_term += 545140134
        exponential_term *= -262537412640768000
        summation += Decimal(numerator_term * linear_term) / exponential_term
        factorial_term_increment += 12

    pi = constant_c / summation
    return str(pi)[:decimal_places + 2]

def main():
    print("Calculate Pi to n decimal places!")
    decimal_places = int(input("Enter the number of decimal places (up to 1000): "))
    if decimal_places > 1000:
        print("Please choose a value up to 1000.")
        return
    pi_value = calculate_pi(decimal_places)
    print(f"Pi to {decimal_places} decimal places:\n{pi_value}")

if __name__ == "__main__":
    main()
```

#### 3. Create a Python project to guess a number that has randomly selected.

```python
import random

def fun():
    random_number = random.randint(1, 100)
    attempts = 0
    print("Guess the number between 1 and 100.")

    while True:
        user_guess = int(input("Enter your guess: "))
        attempts += 1

        if user_guess < random_number:
            print("Too low. Try again.")
        elif user_guess > random_number:
            print("Too high. Try again.")
        else:
            print(f"Congratulations! You guessed the number in {attempts} attempts.")
            break

fun()
```

#### 8. Write a Python program to list countries with all cases of Novel Coronavirus (COVID-19) recovered.

```python
import requests
from bs4 import BeautifulSoup

def counteries_with_full_recovery():
    url = "https://www.worldometers.info/coronavirus/"
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table', {'id': 'main_table_countries_today'})
        
        countries_data = []
        rows = table.find_all('tr')[1:]  # skiping the header row
        
        for row in rows:
            columns = row.find_all('td')
            if len(columns) < 8:  # Skip rows that don't have enough columns
                continue
            country_name = columns[1].text.strip()
            try:
                # here the indexing is correct, but on the https://www.worldometers.info/coronavirus/, some columns are hiddent, so it was bit confucing for the first time
                total_cases = int(columns[2].text.strip().replace(",", ""))
                total_recovered_text = columns[6].text.strip().replace(",", "")
                
                # Skip rows where 'Total Recovered' is 'N/A' or invalid
                # total_recovered_text is not directly converted to the integer, because of carrying out following comparison
                if total_recovered_text == 'N/A' or not total_recovered_text.isdigit():
                    continue
                
                total_recovered = int(total_recovered_text)

                # this if conditions, determines, wether the total_cases are same as total_recovered_cases
                # and also checks wether there was any case for the same or not
                # as if there is no case, there is no recoveries
                if total_cases == total_recovered and total_cases > 0:
                    countries_data.append(country_name)
            except ValueError:
                # Skip rows with invalid or missing data
                continue
        
        if countries_data:
            print("Countries with all COVID-19 cases recovered:")
            for country in countries_data:
                print(country)
        else:
            print("No countries currently have all cases fully recovered.")
    else:
        print("Failed to fetch data from Worldometer.")

counteries_with_full_recovery()
```

#### 7. Write a Python program to list countries with all cases of Novel Coronavirus (COVID-19) died

```python
import requests
from bs4 import BeautifulSoup

def countries_with_all_corona_patient_dead():
    url = "https://www.worldometers.info/coronavirus/"
    response = requests.get(url)

    if(response.status_code == 200):
        soup = BeautifulSoup(response.text, "html.parser")
        table = soup.find('table', {'id': 'main_table_countries_today'})

        countries_data = []
        rows = table.find_all('tr')[1:]

        for row in rows:
            columns = row.find_all('td')
            if(len(columns) < 8):
                continue
            country_name = columns[1].text.strip()
            try:
                total_cases = int(columns[2].text.strip().replace(',', ''))
                total_death_text = columns[4].text.strip().replace(',', '')
                if not total_death_text.isdigit():
                    continue
                total_deaths = int(total_death_text)

                if total_deaths == total_cases and total_cases > 0:
                    countries_data.append(country_name)
            except ValueError:
                continue

        if countries_data:
            for country in countries_data:
                print("Countries with equal cases and death are: ")
                print(country)
        else:
            print("There are no such countries having same deaths and cases")
    else:
        print("cant fetch results from worldometer")

countries_with_all_corona_patient_dead()
```

#### 5. Write a Python program to get the latest country wise deaths cases of Novel Coronavirus (COVID-19).

```python
# Solution 2
import requests
from bs4 import BeautifulSoup

def fetch_coronavirus_data():
    url = "https://www.worldometers.info/coronavirus/"
    response = requests.get(url)

    if(response.status_code == 200):
        countries_data = []
        
        # Parse the HTML content
        soup = BeautifulSoup(response.text, "html.parser")
        table = soup.find('table', {'id': 'main_table_countries_today'})

        # Fetch rows in the table, excluding the header
        rows = table.find_all('tr')[1:]  # Skipping the first row (header)

        # Define a list of non-country entities to exclude (like continents, world, etc.)
        non_country_entities = ["World", "North America", "Asia", "Europe", "Africa", "Oceania", "South America"]

        for row in rows:
            columns = row.find_all('td')
            if len(columns) < 7:  # Skip rows with insufficient data
                continue

            # Extract data from each column
            country_name = columns[1].text.strip()
            
            # Skip rows that represent continents or world
            if country_name in non_country_entities:
                continue

            total_cases = columns[2].text.strip().replace(",", "")
            total_deaths = columns[4].text.strip().replace(",", "")
            total_recovered = columns[6].text.strip().replace(",", "")

            # Save the extracted data in a dictionary
            countries_data.append({
                'country': country_name,
                'total_cases': total_cases,
                'total_deaths': total_deaths,
                'total_recovered': total_recovered
            })

        # Output the extracted data
        for country_data in countries_data:
            print(f"Country: {country_data['country']}")
            print(f"Total Cases: {country_data['total_cases']}")
            print(f"Total Deaths: {country_data['total_deaths']}")
            print(f"Total Recovered: {country_data['total_recovered']}")
            print("-" * 40)
    else:
        print("Failed to fetch data from Worldometer")

# Call the function to fetch and display data
fetch_coronavirus_data()
```

#### 2. Write a Python program to get the latest number of confirmed, deaths, recovered and active cases of Novel Coronavirus (COVID-19) Country wise.

```python
import pandas as pd
covid_data= pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-17-2020.csv')
covid_data['Active'] = covid_data['Confirmed'] - covid_data['Deaths'] - covid_data['Recovered']
result = covid_data.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().reset_index()
# .sum will sum the repeated values with same group by attribute, and reset index will reset the index for the same
print(result) 
```
